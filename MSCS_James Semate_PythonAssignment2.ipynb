```python
#Instruction: Submit a GitHub repository link and a well-documented Jupyter Notebook (.ipynb) file and/or .py file(s) on Canvas or Moodle. Ensure the notebook runs without errors and includes clear explanations, code, and outputs.
#Q1: Electricity Billing Optimization
#You are analyzing monthly household electricity consumption data for 10 customers in Kampala. Each record contains: customer_name, units_used, connection_type (e.g., 'Bosco', 245, 'Domestic').
customers = [
    ('Bosco', 245, 'Domestic'),
    ('Jane', 400, 'Commercial'),
    ('Peter', 120, 'Domestic'),
    ('Sarah', 600, 'Commercial'),
    ('John', 520, 'Domestic'),
    ('Mary', 310, 'Domestic'),
    ('Paul', 700, 'Commercial'),
    ('Rita', 480, 'Domestic'),
    ('David', 800, 'Commercial'),
    ('Agnes', 150, 'Domestic')
]
#Attempt the following tasks:

#1.1. Use zip() and map() to pair customer names with total bills computed by a lambda: 
#  Domestic rate = 820 UGX/unit
# Commercial rate = 1000 UGX/unit
# Extract names,units,types separately in different lists
names = [record[0] for record in customers]
units = [record[1] for record in customers]
types = [record[2] for record in customers]
#map() computes each customer’s bill based on their connection type, and zip() pairs each name with the corresponding bill
# Compute total bills using map() and a lambda function
bills = list(map(lambda record: 
                 record[1] * (820 if record[2] == 'Domestic' else 1000), 
                 customers))

# Pair names with their bills using zip()
name_bill_pairs = list(zip(names, bills))

print("Names paired with their respective total bills (UGX):")
print(name_bill_pairs)
print()

#1.2. Use filter() to separate high-consumption customers (units > 500).
#

high_consumers = list(filter(lambda record: record[1] > 500, customers))
print("High-consumption customers (units > 500):")
for c in high_consumers:
    print(c)
print()

#1.3. Using *args, write a function average_bill(*bills) that computes the average.
#*args Accepts a variable number of bill amounts and returns their average.
def average_bill(*bills):
    """Compute the average of all provided bills."""
    if len(bills) == 0:  #gets the number of elements in bills and they must not be equal  0
        return 0
    return sum(bills) / len(bills)

avg = average_bill(*bills)
print(f"Average bill: UGX {avg:,.0f}")
print()


#1.4. Display results using list comprehension in the format: ['Bosco: UGX 200,900', 'Jane:UGX 400,000', ...]
#List comprehension Formats each record into a neat string like 'Bosco: UGX 200,900'.
formatted_bills = [f"{name}: UGX {bill:,.0f}" for name, bill in name_bill_pairs]

print("Formatted bill list:")
print(formatted_bills)
print()
```

```python
############################################################################################
print("Q2: Market Basket Price Aggregator")
# Each vendor reports different fruit prices daily:

mango_prices = [2500, 2700, 2600, 2800] 
orange_prices = [3000, 3200, 3100, 3050] 
apple_prices = [4500, 4600, 4550, 4700]
# Attempt the following tasks:
# 2.1 Use zip_longest() to handle missing prices.
#     This ensures that if one list is shorter, missing values are filled with None
all_prices = list(zip_longest(mango_prices, orange_prices, apple_prices, fillvalue=None))
print("Combined Vendor Price List (with None for missing data):")
for i, prices in enumerate(all_prices, start=1):
    print(f"Vendor {i}: Mango={prices[0]}, Orange={prices[1]}, Apple={prices[2]}")
#print("-" * 50)


# 2.2 Define a lambda function avg = lambda lst: sum(lst)/len(lst) to compute average per fruit.
avg = lambda lst: sum(lst) / len(lst)  #lambda is simply a short function

# Calculate average price for each fruit
mango_avg = avg(mango_prices)
orange_avg = avg(orange_prices)
apple_avg = avg(apple_prices)

# Store averages in a dictionary for easy reference
fruit_averages = {
    "Mango": mango_avg,
    "Orange": orange_avg,
    "Apple": apple_avg
}

print("\nAverage Prices:")
for fruit, price in fruit_averages.items():
    print(f"{fruit}: {price:.2f}")
print("-" * 50)

# 2.3 Create a generator expression that yields all fruits with average price above 3000.
above_avg_gen = (fruit for fruit, price in fruit_averages.items() if price > 3000)

# 2.4 Print the generator’s results neatly as: Fruits Above Average Price: Mango, Orange, Apple
above_avg_fruits = list(above_avg_gen)  # Convert generator to list for printing
print("Fruits Above Average Price: " + ", ".join(above_avg_fruits))

```

```python
print("Q3: District Temperature Tracker")
#Temperature data (°C):
kampala = [28, 30, 29, 27, 26, 29, 30, 31, 32, 30, 29, 28] 
gulu = [25, 26, 27, 27, 28, 29, 30, 30, 29, 27, 26, 25] 
mbarara = [22, 23, 23, 24, 25, 25, 26, 27, 27, 26, 24, 23]
#Attempt the following tasks:
 
print("3.1. Use map() with lambda to convert all temperatures to Fahrenheit.")
""" MAP applys a specified function to each item in an iterable (like a list, tuple, or set) and returns a new iterable"""
kampala_f = list(map(lambda c: (c * 9/5) + 32, kampala))
gulu_f = list(map(lambda c: (c * 9/5) + 32, gulu))
mbarara_f = list(map(lambda c: (c * 9/5) + 32, mbarara))

print("Kampala Temperatures in Fahrenheit:")
print(kampala_f)
print()

print("3.2. Use zip() to pair months (Jan–Dec) with Kampala temps.")
# List of months for pairing
months = ["January", "February", "March", "April", "May", "June",
          "July", "August", "September", "October", "November", "December"]


kampala_monthly = list(zip(months, kampala))
print("Kampala Monthly Temperatures (°C):")
for month, temp in kampala_monthly:
    print(f"{month}: {temp}°C")
print()
print("3.3. Use filter() to identify months with temperature > 30°C.")

hot_months = list(filter(lambda m_t: m_t[1] > 30, kampala_monthly))
print("Months in Kampala with temperature > 30°C:")
for month, temp in hot_months:
    print(f"{month}: {temp}°C")
print()
print("3.4. Create a generator to yield month names and temperatures above threshold.")

def hot_months_generator(city_data, threshold=30):
    """Yields month and temperature if above threshold."""
    for month, temp in city_data:
        if temp > threshold:
            yield month, temp
            
            
hot_gen = hot_months_generator(kampala_monthly, 30)
hot_list = [f"{month} - {temp}°C" for month, temp in hot_gen]
print("3.5. Print as: Hot months in Kampala: July - 31°C, August - 32°C")
if hot_list:
    print("Hot months in Kampala:", ", ".join(hot_list))
else:
    print("No months in Kampala above 30°C.")
```

```python
print("Q4: School Fees Payment Analyzer")
# Installment data
students = ['Alex', 'Grace', 'Sarah', 'Brian']
installments = [
    [150000, 200000, 250000],   # Alex's payments
    [500000, 0, 200000],        # Grace's payments
    [300000, 300000, 300000],   # Sarah's payments
    [400000, 100000, 0]         # Brian's payments
]


print("4.1. Lambda function to sum valid payments (ignore zeros)")
# The lambda filters out zeros and sums only non-zero payments
sum_valid_payments = lambda payments: sum(p for p in payments if p > 0)
#print("sum valid payments:",sum_valid_payments())

print("4.2. Use map() to compute total paid by each student")
# map() applies the lambda function to every sub-list in installments
total_paid = list(map(sum_valid_payments, installments))


print("4.3. Combine student names and totals using zip()")

student_totals = list(zip(students, total_paid))

# Display student totals
print("Total Payments by Students")
for name, total in student_totals:
    print(f"{name}: {total:,} UGX")


print("4.4. Use filter() to find students who have cleared full fees (≥ 600,000)")
cleared_students = list(filter(lambda item: item[1] >= 600000, student_totals))

print("\nStudents Who Have Cleared Full Fees (≥600,000)")
if cleared_students:
    for name, total in cleared_students:
        print(f"{name}: {total:,} UGX")
else:
    print("No student has fully cleared their fees.")


print("4.5. Define payment_summary(**kwargs)")

def payment_summary(**kwargs):
    """Receives named arguments like Alex=600000 and prints summaries."""
    print("\nPayment Summary Report")
    for name, amount in kwargs.items():
        status = "Cleared" if amount >= 600000 else "Pending"
        print(f"Student: {name}, Paid: {amount:,} UGX, Status: {status}")


# Call the payment_summary function
# Unpack the student_totals list into named arguments using dict()
payment_summary(**dict(student_totals))
```

```python

print("Q5: Agricultural Yield Estimator")
#Yield data:
districts = ['Bushenyi', 'Mityana', 'Kasese', 'Mbale']
yield_data = [1200, 1500, 900, 1300]  # in kilograms (kg)

print("5.1: Use list comprehension with a lambda to convert yields from kg to tons")
# 1 ton = 1000 kg
# We use a lambda inside a list comprehension to divide each yield by 1000
"""A lambda function can take any number of arguments, but can only have one expression"""
yields_in_tons = [(lambda y: y / 1000)(y) for y in yield_data]
print("Yields in tons:", yields_in_tons)


print("5.2: Write a generator that yields only districts with yield > 1 ton ")
def high_yield_districts(districts, yields):
    """Generator that yields (district, yield_in_tons) for yields > 1 ton"""
    for d, y in zip(districts, yields):  #zip combines the two equal length lists
        if y > 1:                        #extracts only yield above 1 ton
            yield (d, y)

print("\nDistricts with yield > 1 ton:")
for district, yield_ton in high_yield_districts(districts, yields_in_tons):
    print(f"{district}: {yield_ton} tons")



print("5.3: Use *args to compute average yield across all districts ")
def average_yield(*args):
    """Compute average of all yield values passed in"""
    return sum(args) / len(args)

avg_yield = average_yield(*yields_in_tons)
print(f"\nAverage yield across all districts: {avg_yield:.2f} tons")

print(" 5.4: Use **kwargs to simulate price changes per district and compute revenue ")
# Let's assume each district has a different price per ton (in UGX)
prices = {
    'Bushenyi': 5_000_000,
    'Mityana': 4_800_000,
    'Kasese': 5_000_000,
    'Mbale': 4_700_000
}

def compute_revenue(yields, districts, **kwargs):
    """
    Compute revenue = yield_in_tons × price_per_ton for each district.
    kwargs simulate changing prices (district=price)
    **kwargs allows a function to accept a variable number of keyworded arguments
    """
    revenues = {}  # empty dictonary
    for d, y in zip(districts, yields):
        price = kwargs.get(d, 0)
        revenues[d] = y * price
    return revenues

revenues = compute_revenue(yields_in_tons, districts, **prices)

print("5.5: Print formatted output")
print("\n District Production and Revenue Report ")
for d, y in zip(districts, yields_in_tons):  #iterates through a combined list
    print(f"{d} produced {y:.1f} tons — Revenue: UGX {revenues[d]:,.0f}")
```

```python

print("Q6: Web Data Aggregation")
#Given that sites = ['https://ucu.ac.ug', 'https://harba.ug', 'https://www.bou.or.ug']
#Attempt the following tasks:
import requests                   # Imported REQUEST library  FOR interacting with web services and APIs
# websites list
sites = ['https://ucu.ac.ug', 'https://harba.ug', 'https://www.bou.or.ug']

print("6.1 function that takes *urls and uses requests.get() to return response codes")
def get_status_codes(*urls):
    """
    This function accepts multiple URLs (using *args)
    and returns a dictionary containing each URL and its HTTP response code.
    If a URL cannot be reached, it assigns code 0.
    """
    results = {}   # create an empty dictionary for url and status_code pairs
    for url in urls:
        try:
            response = requests.get(url, timeout=6)
            results[url] = response.status_code
        except requests.RequestException as e:
            # Handles exceptions  such as ConnectionError, Timeout, etc.
            results[url] = 0
            print(f"Error accessing {url}: {e}")
    return results


# Call the function with the list of sites
""" function to accept a variable number of positional arguments"""
response_codes = get_status_codes(*sites) # argument *sites is passed in the function get_status_codes to get a dictionary of url:codes

# Print all response codes
print("\n#### All Site with their respective Response Codes ####")
for url, code in response_codes.items():
    print(f"{url} : {code}")

print("6.2 use list comprehension to print only URLs with status code 200")
active_sites = [url for url, code in response_codes.items() if code == 200] #filters out urls with code 200 in the iterable list

print("\n#### Active Sites (Status 200) ####")
print(active_sites)

print("6.3 Store results in a dictionary using dictionary comprehension")
site_status_dict = {url: code for url, code in response_codes.items()} #creates key:value pairs for url:codes
print("\n#### Dictionary of Site Status Codes ####")
print(site_status_dict)

print("6.4 Create a generator expression that yields 'Active Site: <url>' for reachable domains")
active_site_generator = (f"Active Site: {url}" for url, code in response_codes.items() if code == 200)#extracts urls when code==200


print("\n#### Active site Generator Output ####")
for message in active_site_generator:
    print(message)
```
